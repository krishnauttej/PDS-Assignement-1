```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('/content/data_clean/StudentsPerformance.csv')
df.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
df.head(10)
```

```python
# Plot 1: Bar Plot of Race/Ethnicity Distribution
plt.figure(figsize=(8, 6))
sns.countplot(x='race_or_ethnicity', data=df, palette='viridis')
plt.title('Race/Ethnicity Distribution')
plt.xlabel('Race/Ethnicity')
plt.ylabel('Count')
plt.show()
```

```python
# Plot 2: Box Plot of Math, Reading, and Writing Scores by Gender
plt.figure(figsize=(10, 6))
sns.boxplot(x='gender', y='math_score', data=df)
sns.boxplot(x='gender', y='reading_score', data=df)
sns.boxplot(x='gender', y='writing_score', data=df)
plt.title('Math, Reading, and Writing Scores by Gender')
plt.xlabel('Gender')
plt.ylabel('Score')
plt.show()
```

```python
# Plot 3: Histogram of Lunch Type
plt.figure(figsize=(8, 6))
sns.countplot(x='lunch', data=df, palette='plasma')
plt.title('Lunch Type Distribution')
plt.xlabel('Lunch Type')
plt.ylabel('Count')
plt.show()
```

```python
# Plot 4: Bar Plot of Average Scores by Parental Level of Education
avg_scores = df.groupby('parental_level_of_education').mean()[['math_score', 'reading_score', 'writing_score']]
plt.figure(figsize=(10, 6))
avg_scores.plot(kind='bar', cmap='Set2')
plt.title('Average Scores by Parental Level of Education')
plt.xlabel('Parental Level of Education')
plt.ylabel('Average Score')
plt.xticks(rotation=45)
plt.legend(title='Score')
plt.show()
```

```python
# Plot 5: Violin Plot of Writing Scores by Race/Ethnicity and Gender
plt.figure(figsize=(10, 6))
sns.violinplot(x='race_or_ethnicity', y='writing_score', hue='gender', data=df, palette='Set2', split=True)
plt.title('Writing Scores by Race/Ethnicity and Gender')
plt.xlabel('Race/Ethnicity')
plt.ylabel('Writing Score')
plt.legend(title='Gender')
plt.show()
```
