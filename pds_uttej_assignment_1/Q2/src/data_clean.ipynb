```python
import pandas as pd
```

```python
df = pd.read_csv('/content/data_raw/StudentsPerformance.csv')
```

```python
 #Drop rows where the value in 'test_preparation_course' column is None
df = df[df['test_preparation_course'] != 'none']
df.reset_index(drop=True, inplace=True)
```

```python
df.head(10)
```

```python
print("\nChecking for missing values:")
print(df.isnull().sum())
```

```python
df.columns = df.columns.str.strip()
```

```python
# Remove duplicates (if any)
df.drop_duplicates(inplace=True)
```

```python
df.rename(columns = {'race/ethnicity':'race_or_ethnicity',
                     'parental level of education':'parental_level_of_education',
    'test preparation course':'test_preparation_course',
                       'math score':'math_score',
                       'reading score':'reading_score',
                        'writing score':'writing_score'}, inplace = True)
```

```python
df.to_csv('/content/data_clean/StudentsPerformance.csv')
```
