```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('/content/data_clean/clean_frality.csv')
```

```python
df.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='frailty', data=df, palette='Set2')
plt.title('Frailty Count')
plt.xlabel('Frailty')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='frailty', y='grip_strength', data=df, hue='frailty', palette='Set1')
plt.title('Grip Strength Distribution across Frailty Categories')
plt.xlabel('Frailty')
plt.ylabel('Grip Strength')
plt.show()
```
